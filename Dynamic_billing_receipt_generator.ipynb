```python
from reportlab.lib.pagesizes import A4
from reportlab.lib import colors
from reportlab.lib.colors import black, white,whitesmoke, gray, yellow
from reportlab.lib.pagesizes import A4
from reportlab.lib.styles import getSampleStyleSheet, ParagraphStyle
from reportlab.platypus import SimpleDocTemplate, Table, TableStyle, Paragraph, Image, Spacer

# Function to create the receipt
def receipt(file_name, items):
    pdf = SimpleDocTemplate(file_name, pagesize=A4)
    elements = []

    # Adding a logo
    logo = Image("logo.png", width=100, height=50)
    elements.append(logo)
    elements.append(Spacer(1, 20))

    # Title
    title = Paragraph("Billing Receipt", ParagraphStyle(name="Title", fontSize=18, textColor=black, alignment=1))
    elements.append(title)
    elements.append(Spacer(1, 20))

    # Header
    headers = ["Item Name", "Quantity", "Price (USD)"]

    # Convert items to a table format
    data = [headers] + [[item['name'], item['quantity'], f"$ {item['price']:,}"] for item in items]

    # Create table
    table = Table(data)
    table.setStyle(TableStyle([
        ('BACKGROUND', (0, 0), (-1, 0), gray),
        ('TEXTCOLOR', (0, 0), (-1, 0), white),
        ('ALIGN', (0, 0), (-1, -1), 'CENTER'),
        ('FONTNAME', (0, 0), (-1, 0), 'Helvetica-Bold'),
        ('FONTSIZE', (0, 0), (-1, -1), 12),
        ('BOTTOMPADDING', (0, 0), (-1, 0), 12),
        ('BACKGROUND', (0, 1), (-1, -1), colors.beige),
        ('GRID', (0, 0), (-1, -1), 0.5, black),
    ]))

    elements.append(table)

    # Spacer for separation
    elements.append(Spacer(1, 20))

    # Calculating total amount
    total_amount = sum(item['quantity'] * item['price'] for item in items)

    # Adding total row outside the table
    total_data = [["", "Total", f"$ {total_amount:,}"]]
    total_table = Table(total_data)
    total_table.setStyle(TableStyle([
        ('BACKGROUND', (1, 0), (-1, 0), whitesmoke),
        ('TEXTCOLOR', (0, 0), (-1, 0), black),
        ('ALIGN', (1, 0), (-1, -1), 'CENTER'),
        ('FONTNAME', (0, 0), (-1, 0), 'Helvetica-Bold'),
        ('FONTSIZE', (0, 0), (-1, -1), 12),
        ('GRID', (1, 0), (-1, 0), 0.5, black),
    ]))

    elements.append(total_table)

    # Build PDF
    pdf.build(elements)

# Main function to take user input and generate the receipt
def main():
    items = []
    while True:
        name = input("Enter item name (or type 'done' to finish): ")
        if name.lower() == 'done':
            break;
        quantity = int(input("Enter quantity: "))
        price = float(input("Enter price (USD): "))
        items.append({'name': name, 'quantity': quantity, 'price': price})

    receipt("receipt.pdf", items)
    print("Receipt generated successfully as 'receipt.pdf'.")

if __name__ == "__main__":
    main()
```
